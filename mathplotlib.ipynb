In questo notebook utilizzerò la libreria Python di visualizzazione dati mathplotlib, e nello specifico il suo modulo pyplot, e la libreria Pandas, per creare una tabella che riporti i dati di vendita dei 10 videogiochi più venduti sul mercato mondiale.

```python
#importo le librerie necessarie
import matplotlib.pyplot as plt
import pandas as pd #mi servirà per il dataframe

#creo il dataframe utilizzando la libreria pandas:
dati = {
    'Vendite' : [238000000, 180000000, 100000000, 82900000, 75000000, 62250000, 58000000, 53000000, 47520000, 44500000],
    'Console' : ["Multi-platform", "Multi-platform", "Multi-platform", "Nintendo Wii", "Multi-platform", "Nintendo Wii U/ Switch", "Multi-platform", "Multi-platform", "Multi-platform", "Multi-platform"],
    'Data di rilascio' : ["November 18, 2011", "September 17, 2013", "September 12, 2006", "November 19, 2006", "December 20, 2017", "May 29, 2014", "September 13, 1985", "October 26, 2018", "February 27, 1996", "May 16, 2011"],
    'Casa di sviluppo' : ["Mojang Studios", "Rockstar North", "EA Mobile", "Nintendo EAD", "PUBG Corporation", "Nintendo EAD", "Nintendo R&D4", "Rockstar Studios", "Game Freak", "Re-Logic"]

}

dataframe = pd.DataFrame(dati, index = ['Minecraft', 'GTA V', 'Tetris(EA)', 'Wii Sports', 'PUBG: Battlegrounds', 'Mario Kart 8 / Deluxe', 'Super Mario Bros.', 'Red Dead Redemption 2', 'Pokémon Red / Green / Blue / Yellow', 'Terraria'])

#ora creo la tabella vera e propria , utilizzando il metodo table() e subplot() del dataframe: 
figura, assi = plt.subplots() #il metodo subplots() restituirà due oggetti: la figura e gli assi del grafico;
tabella = assi.table(cellText=dataframe.values, colLabels=dataframe.columns, rowLabels=dataframe.index, loc='center')#specifico le etichette di righe e colonne della tabella, e la posiziono al centro della figura;

#imposto lo stile della tabella:
tabella.auto_set_font_size(False) #determino la deminesione del font
tabella.set_fontsize(14)
tabella.scale(2, 4) #determino la dimensione della tabella

#nascondo gli assi per motivi estetici
assi.axis('off')


#ora posso mostrare la tabella, grazie al metodo show() di matplotlib.pyplot:
print("Videogames più venduti a livello globale:\n")
plt.show()
```
